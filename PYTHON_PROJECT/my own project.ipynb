```python
from flask import Flask, render_template, request,jsonify
from flask_cors import CORS,cross_origin
import requests
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq

searchString="sumsung"
flipkart_url = "https://www.flipkart.com/search?q="+ searchString
flipkart_url

httpObject=uReq(flipkart_url)
webdata=httpObject.read()
#print(webdata)

soupdata=soup(webdata,'html.parser')




href=soupdata.find_all('a',{'class':'_1fQZEK'})
domain='https://www.flipkart.com'

productLink=domain+href[0].get('href')

prodRes = requests.get(productLink)

prodRes.encoding='utf-8'

prod_html = soup(prodRes.text, "html.parser")


ratings=prod_html.find_all('div',{'class':'_3LWZlK _1BLPMq'})
for i in range(len(ratings)):
    print(prod_html.find_all('div',{'class':'_3LWZlK _1BLPMq'})[i].text)
    
    
coment_tag=prod_html.find_all('p',{'class':'_2-N8zT'})
for i in range(len(coment_tag)):
    print(coment_tag[i].text)
    
    
Names=prod_html.find_all('p',{'class':'_2sc7ZR _2V5EHH'})
for i in range(len(Names)):
    print(Names[i].text)    
```
